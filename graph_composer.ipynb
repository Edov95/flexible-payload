```python
import matplotlib.pylab as plt
import matplotlib
```

```python
training_1 = []

# open file and read the content in a list
with open('training_1.txt', 'r') as filehandle:
    for line in filehandle:
        # remove linebreak which is the last character of the string
        currentPlace = float(line[:-1])

        # add item to the list
        training_1.append(currentPlace)
```

```python
matplotlib.rcParams.update({'font.size': 14})

plt.figure(figsize=(20,4))
plt.plot(training_1, color='red', label='Training 1')
plt.plot(training_2, color='purple', label='Training 2')
plt.plot(training_3, color='grey', label='Training 3')
plt.plot(training_4, color='green', label='Training 4')
plt.plot(training_5, color='blue', label='Training 5')
# plt.scatter(np.arange(len(rew)), rew, color='red', label='Actions dones')
# plt.bar(np.arange(len(simulation)), reward, color='blue', label='Reward')
plt.title('Mean reward for each training')
plt.grid()
plt.legend()
#plt.savefig('Results')
#plt.xlim([0, 100])
plt.xlabel("Scenario")
plt.ylabel("Reward")
plt.ylim([-3, 0.5])
plt.show()
```
